```python
from functools import reduce #reduce my beloved S2(Utilizado ao invÃ©s de 'for' quando conveniente)
from pathlib import Path #usado para encontrar o path do arquivo no computador na hora de exportar o csv
import pandas as pd #usado para manipular dados de maneira mais eficiente
import requests #biblioteca de webscrap
import re #usado para localizar dados especificos da request
from bs4 import BeautifulSoup #usado para localizar dados especificos do html
from datetime import datetime #usado para fazer as timestamps

#funÃ§Ã£o usa reduce para transformar conjunto de soup.text em uma list do python
def ListBuilder(x,y):
    if type(x) != list:
        return [x.text.strip().rstrip(':')]+[y.text.strip().rstrip(':')]
    else:
        return x+[y.text.strip().rstrip(':')]
    
#funÃ§Ã£o usa reduce para fazer o merge de todos os paises em um DataFrame pandas
def MegaMerge(x,y):
    return x.merge(y,how='outer')

#Aqui ocorre o webscrap do sitemap
response = requests.get(url="http://127.0.0.1:8000/places/default/sitemap.xml")
xml = response.content.decode()
urlcode = re.findall("<loc>(.*?)<", xml)

#Listas vazias sÃ£o criadas para armazenar futuros dados na memÃ³ria
tempinfo = {}
infoList = []

#for pega a url de cada paÃ­s do sitemap e cria um Dict com as keys e informaÃ§Ãµes de cada um e entÃ£o
#um DataFrame para cada paÃ­s que Ã© posto em uma lista
for i in urlcode:
    page = requests.get(url=i)
    decode = page.content.decode()
    soup = BeautifulSoup(decode)
    listaContent = soup.body.table.find_all(class_="w2p_fw")
    listaKeys = soup.body.table.find_all(class_="w2p_fl")
    contents = reduce(ListBuilder,listaContent)
    contents.remove('')#Ã© removido da lista de conteÃºdos o '.text' da bandeira(vazio)
    contents.insert(0,listaContent[0].img)#Ã© inserido no lugar o img src dela
    keys = reduce(ListBuilder,listaKeys)
    obj = str(datetime.now())
    contents.append(obj)
    keys.append('Timestamp')
    for key in keys:
        for content in contents:
            tempinfo[key] = content
            contents.remove(content)
            break
    info = pd.DataFrame([tempinfo])
    infoList.append(info)

    
#Por fim a lista Ã© reduzida para um Ãºnico DataFrame com todos os paÃ­ses e exportado para um arquivo csv
final = reduce(MegaMerge,infoList)
display(final)
final.to_csv(str(Path.cwd())+'/ListaDePaises.csv')
```

```python
#Pode por o cÃ³digo pra rodar e ir tomar um suco, vai demorar
from functools import reduce #reduce my beloved S2(Utilizado ao invÃ©s de 'for' quando conveniente)
import requests #biblioteca de webscrap
import re # "Some people when confronted with a problem, think 'I know, I'll use regular expressions.'
          # Now they have two problems." Jamie Zawinski
import pandas as pd #pandas Ã© muito bom para manipular e exportar dados
from pathlib import Path #usado para encontrar o path do arquivo no computador na hora de exportar o json
from bs4 import BeautifulSoup #usado para localizar dados especificos do html

#funÃ§Ã£o usa reduce para fazer o merge de todos os [-p-a-i-s-e-s-] filmes em um DataFrame pandas
def MegaMerge(x,y):
    return x.merge(y,how='outer')


site = 'https://www.imdb.com'
page = requests.get(url='https://www.imdb.com/chart/top/')
decode = page.content.decode()
soup = BeautifulSoup(decode)

info = []#Lista para salvar a informaÃ§Ã£o

#filme Ã© a lista de links dos 250 top filmes
filme = soup.tbody.find_all(class_="titleColumn")
for i in filme:#for transforma i em cada um dos links
    split = str(i.a).split('\"')
    link = split[1]#Aqui o link Ã© isolado
    #request+decode
    page = requests.get(url=site+link,headers={'user-agent':'Mozilla/5.0'})
    decode = page.content.decode()
    soup = BeautifulSoup(decode)
    
    #InformaÃ§Ãµes requisitadas(nÃ£o consegui achar o genero explicaÃ§Ã£o abaixo)
    title = soup.body.main.find_all(class_="sc-afe43def-1 fDTGTb")[0].text
    score = soup.find(class_="sc-bde20123-1 iZlgcd").text
    ano = soup.find_all(class_="ipc-link ipc-link--baseAlt ipc-link--inherit-color")[4].text
    posterUrl = soup.find(class_='ipc-image')['src']
    posterImg = soup.find(class_="ipc-lockup-overlay ipc-focusable")['href']
    diretores = soup.find(class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link").text
    tempInfo = {
        'TÃ­tulo':title,
        'Nota':score,
        'Ano':ano,
        'Url do poster':posterUrl,
        'Imagem do poster':posterImg,
        'Diretores':diretores,
    }
    filmInfo = pd.DataFrame([tempInfo])
    info.append(filmInfo)
    
    
#Merge+export
final = reduce(MegaMerge,info)
display(final)
final.to_json(str(Path.cwd())+'/ListaDeFilmes.json')
```
