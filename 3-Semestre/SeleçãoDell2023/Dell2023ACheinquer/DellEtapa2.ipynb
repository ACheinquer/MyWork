```python
########imports
from pathlib import Path
import re
import pandas as pd
import math
from tkinter import*


########Classes que armazenam a informação do programa
class Trecho: #Aqui serão armazenadas informações individuais de cada trecho de uma viagem
    __slots__=['nome','IdViagem','nViagem','Km','custo','P','M','G']
    def __init__(self,nome,IdViagem,nViagem,Km,custo,P,M,G):#No código chamo 1 transporte inteiro de viagem
        self.nome = nome #nome do trecho
        self.IdViagem = IdViagem #numero de id da viagem
        self.nViagem = nViagem #numero de viagens realizadas no trecho
        self.Km = Km #kilometragem total do trecho
        self.custo = custo #custo total do trecho
        self.P = P#caminhões pequenos utilizados
        self.M = M#caminhões medios utilizados
        self.G = G#caminhões grandes utilizados

class ItemInfo:#Aqui serão armazenadas informações especificas dos itens
    __slots__=['nome','peso']
    def __init__(self,nome,peso):
        self.nome = nome#nome do item
        self.peso = peso #peso individual do item
        
class Caminhao:
    __slots__=['tipo','custo','limite']
    def __init__(self,tipo,custo,limite):
        self.tipo = tipo
        self.custo = custo
        self.limite = limite
        
######## FUNÇÕES DO MENU
def Welcome():
    def Consulta():
        janela_viagem = Tk()
        janela_viagem.title('Consulta')
        def DoDistance():
            try:
                dist = list(pdCity.loc[cidades.index(partida.get().upper())])[0].split(';')[cidades.index(destino.get().upper())]
                if rawmodel.get().lower() == pequeno.tipo:
                    result = float(dist)*pequeno.custo
                elif rawmodel.get().lower() == medio.tipo:
                    result = float(dist)*medio.custo
                elif rawmodel.get().lower() == grande.tipo:
                    result = float(dist)*grande.custo
                else:
                    raise ValueError()
                show_result = Label(janela_viagem,text=50*' ')
                show_result['text']="A distância é "+dist+" e o custo com um caminhão de modalidade "+rawmodel.get()+" será R${:.2f}".format(result)
                show_result.grid(column=1,row=12,pady=(10,5))
            except:
                gram_Error = Label(janela_viagem, text="O nome da(s) cidade(s) ou da modalidade foi escrito errado ou não está disponível para transporte \npor favor consulte a lista e tente novamente \n OBS: escreva Médio sem acento")
                gram_Error.grid(column=1,row=10)
    
        back = Button(janela_viagem,text='Aperte para voltar ao menu principal',command = janela_viagem.destroy)
        back.grid(row=0,column=0)
        City_Select = Label(janela_viagem, text = 'Selecione 2 Cidades entre as seguintes opções:')
        City_Select.grid(column=1,row=0)
        city_options= Label(janela_viagem, text = str(cidades[0:5]))
        city_options1= Label(janela_viagem, text = str(cidades[5:11]))
        city_options2= Label(janela_viagem, text = str(cidades[11:17]))
        city_options3= Label(janela_viagem, text = str(cidades[17:]))

        city_options.grid(column=1,row=1)
        city_options1.grid(column=1,row=2)
        city_options2.grid(column=1,row=3)
        city_options3.grid(column=1,row=4)
    
    
        insert_partida = Label(janela_viagem, text="Insira a cidade de partida")
        insert_partida.grid(column=1,row=5)
        partida = Entry(janela_viagem,width=30)
        partida.grid(column=2, row =5)

        insert_destino = Label(janela_viagem, text="Insira a cidade de destino")
        insert_destino.grid(column=1,row=6)
        destino = Entry(janela_viagem,width=30)
        destino.grid(column=2, row=6)
        
        insert_model = Label(janela_viagem, text="Insira a modalidade de caminhão(Pequeno/Medio/Grande)")
        insert_model.grid(column=1,row=8)
        rawmodel = Entry(janela_viagem,width =30)
        rawmodel.grid(column=2,row=8)
        done = Button(janela_viagem,text='Calcular custo',command=DoDistance)
        done.grid(column=1, row=11)
        janela_viagem.mainloop()
        
    def Cadastrar():
        def CadastroItem():
            def doneTrip():
                cadastro_item.destroy()
                fim  = Label(janela_cadastro,text='O custo total da viagem foi R${:.2f}'.format(custoViagem[0])+'\n Para realizar outra viagem volte ao Menu Principal e abra outra janela')
                fim.grid(column=1,row=11)
                custoViagem.append(custoViagem[0])#viagem atual é salva na lista
                custoViagem[0] = 0#e a primeira posição é zerada para reuso
                nViagem[0]+=1#O número de viagens no código aumenta em 1
                
                
            dist = list(pdCity.loc[cidades.index(cityEntry.get().split('-')[0].upper())])[0].split(';')[cidades.index(cityEntry.get().split('-')[1].upper())]
            itens = re.split(r':|;',itemEntry.get())
            TrechoNome = cityEntry.get().upper()
            pesoTotal = 0
            pesocel = 0
            pesogel = 0
            pesofre = 0            
            pesocad = 0
            pesolum = 0
            pesolav = 0
            for x in range(len(itens)):#Comparação dos itens com os objetos
                if itens[x].lower() == celular.nome:
                    celular.qntTotal = celular.qntTotal+int(itens[x+1])
                    pesocel = celular.peso*int(itens[x+1])
                    
                elif itens[x].lower() == geladeira.nome:
                    geladeira.qntTotal = geladeira.qntTotal+int(itens[x+1])
                    pesogel = geladeira.peso*int(itens[x+1])
                    
                elif itens[x].lower() == freezer.nome:
                    freezer.qntTotal = freezer.qntTotal+int(itens[x+1])
                    pesofre= freezer.peso*int(itens[x+1])
            
                elif itens[x].lower() == cadeira.nome:
                    cadeira.qntTotal = cadeira.qntTotal+int(itens[x+1])
                    pesocad = cadeira.peso*int(itens[x+1])
                        
                elif itens[x].lower() == luminaria.nome:
                    luminaria.qntTotal=luminaria.qntTotal+int(itens[x+1])
                    pesolum =luminaria.peso*int(itens[x+1])
                    
                elif itens[x].lower() == lavadora.nome:
                    lavadora.qntTotal=lavadora.qntTotal*int(itens[x+1])
                    pesolav=lavadora.peso*int(itens[x+1])
                    
            
                
            pesoTotal =  pesocel +pesogel + pesofre + pesocad + pesolum +pesolav
            G = pesoTotal//grande.limite #número de caminhões grandes minimamente necessários
            M = pesoTotal%grande.limite//medio.limite #número de caminhões medios minimamente necessários
            P = pesoTotal%grande.limite%medio.limite//pequeno.limite #número de caminhões pequenos minimamente necessários
            No0G = (pesoTotal%grande.limite>0)
            No0M = (pesoTotal%medio.limite>0)
            No0P = (pesoTotal%pequeno.limite>0)
            
            restoG = pesoTotal%grande.limite
            restoM = pesoTotal%medio.limite
            restoP = pesoTotal%pequeno.limite
            
            
            #aqui compensamos pela divisão deixar resto
            if No0G and No0M and No0P and restoG<pesoTotal and not restoM<=pesoTotal and not restoP<=pesoTotal:
                G = G+1
        
            elif No0G and No0M and No0P and restoM<=pesoTotal and not restoP<=pesoTotal:
                M = M+1
                
            elif No0G and No0M and No0P and restoP<=pesoTotal:
                P = P+1
            
        
            custoTrecho = 0
            dist = int(dist)
            if G>=1 and M<1 and P<1:
                custoTrecho = G*grande.custo*dist
                uso = str(G)+' caminhão(ões) grande(s) com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G>=1 and M>=1 and P<1:
                custoTrecho = (G*grande.custo+M*medio.custo)*dist
                uso = str(G)+' caminhão(ões) grande(s) e '+str(M)+' médios com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G>=1 and M>=1 and P>=1:
                custoTrecho = (G*grande.custo+M*medio.custo+P*pequeno.custo)*dist
                uso = str(G)+' caminhão(ões) grande(s),'+str(M)+' médio(s) e '+str(P)+' pequeno(s) com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G>=1 and M<1 and P>=1:#Caso onde precisa-se de grande(s) e pequeno(s), mas não médios
                custoTrecho= (G*grande.custo+P*pequeno.custo)*dist
                uso = str(G)+ ' caminhão(ões) grandes e '+str(P)+ ' pequeno(s) com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G<1 and M>=1 and P<1:
                custoTrecho = M*medio.custo*dist
                uso = str(M)+' caminhão(ões) medios com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G<1 and M>=1 and P>=1:
                custoTrecho = (M*medio.custo+P*pequeno.custo)*dist
                uso = str(M)+' caminhão(ões) medio(s) e '+str(P)+' pequeno(s) com custo total de R${:.2f}'.format(custoTrecho)
            
            elif G<1 and M<1 and P>=1:
                custoTrecho= P*pequeno.custo*dist
                uso = str(P)+ ' caminão(ões) pequeno(s) com custo total de R${:.2f}'.format(custoTrecho)
            
            custoViagem[0] += custoTrecho#custo desta viagem
            trecho_novo = Trecho(TrechoNome,nViagem[0],1,int(dist),custoTrecho,P,M,G)
            trechos.append(trecho_novo)
                
            donetext = Label(janela_cadastro,text=uso)
            donetext.grid(column=1,row=8)
            inquiry = Label(janela_cadastro,text="Cadastro efetuado, se deseja adicionar mais trehos à viagem repita o processo \n Para finalizar a viagem, aperte o botão abaixo:")
            inquiry.grid(column =1,row=9)
            donebutton = Button(janela_cadastro, text='Aperte para finalizar a viagem',command = doneTrip)
            donebutton.grid(column=1,row=10)
            
        janela_cadastro = Tk()
        janela_cadastro.title("Cadastro")
        back = Button(janela_cadastro,text='Aperte para voltar ao menu principal',command = janela_cadastro.destroy)
        back.grid(row=0,column=0)
        City_Select = Label(janela_cadastro, text = 'Selecione 2 Cidades entre as seguintes opções:')
        City_Select.grid(column=1,row=0)
        city_options= Label(janela_cadastro, text = str(cidades[0:5]))
        city_options1= Label(janela_cadastro, text = str(cidades[5:11]))
        city_options2= Label(janela_cadastro, text = str(cidades[11:17]))
        city_options3= Label(janela_cadastro, text = str(cidades[17:]))
        
        city_options.grid(column=1,row=1)
        city_options1.grid(column=1,row=2)
        city_options2.grid(column=1,row=3)
        city_options3.grid(column=1,row=4)
            
        insert_city = Label(janela_cadastro, text='Insira as 2 cidades do trecho(cidade1-cidade2)')
        insert_city.grid(column=1,row=5)
        cityEntry = Entry(janela_cadastro,width=30)
        cityEntry.grid(column=2,row=5)
        insert_city = Label(janela_cadastro, text='Insira os itens e suas quantidades(item:qnt;item:qnt...)')
        insert_city.grid(column=1,row=6)
        itemEntry = Entry(janela_cadastro,width=30)
        itemEntry.grid(column=2,row=6)
        
        cadastro_item = Button(janela_cadastro,text="Clique aqui para efetivar o cadastro de um item em um trecho \n Podem ser feitos múltiplos cadastros para adicionar mais paradas",command=CadastroItem)
        cadastro_item.grid(column = 1,row=7)
        
        janela_cadastro.mainloop()
        
    def Stats():
        janela_stat = Tk()
        janela_stat.title('Estatísticas')
        
        back = Button(janela_stat,text='Aperte para voltar ao menu principal',command = janela_stat.destroy)
        back.pack()
        
        for x in range(len(custoViagem)-1):
            full_stat = Label(janela_stat,text = 'Custo do '+str(x+1)+'º transporte: R${:.2f}'.format(custoViagem[x+1]))
            full_stat.pack()
            fullKm = 0
            PMG = 0
            Pcost =0
            Mcost =0
            Gcost =0 
            for y in trechos:
                if y.IdViagem == x+1:
                    fullKm += y.Km
                    Pcost += y.P*y.Km*pequeno.custo#calcula o custo para cada modalidade P, M e G do transporte
                    Mcost += y.M*y.Km*medio.custo
                    Gcost += y.G*y.Km*grande.custo
                    PMG += (y.P+y.M+y.G)#calcula a quantidade de caminhões deslocados de cada modalidade
                    trecho_stat = Label(janela_stat,text = 'Custo Trecho '+y.nome+': R${:.2f}'.format(y.custo))
                    trecho_stat.pack()
            
            km_stat = Label(janela_stat,text = 'Custo médio por Km do '+str(x+1)+'º transporte: R${:.2f}'.format(custoViagem[x+1]/fullKm))
            km_stat.pack()                                          
            stat_modP = Label(janela_stat,text = 'Custo dos caminhões pequenos do '+str(x+1)+'º transporte: R${:.2f}'.format(Pcost))
            stat_modP.pack()
            stat_modM = Label(janela_stat,text = 'Custo dos caminhões medios do '+str(x+1)+'º transporte: R${:.2f}'.format(Mcost))
            stat_modM.pack()
            stat_modG = Label(janela_stat,text = 'Custo dos caminhões grandes do '+str(x+1)+'º transporte: R${:.2f}'.format(Gcost))
            stat_modG.pack()
            stat_PMG = Label(janela_stat,text = 'Número total de caminhões deslocados no '+str(x+1)+'º transporte: '+str(PMG))
            stat_PMG.pack()
                              
        janela_stat.mainloop()
        
        
    #Código da função welcome
    welcome = Label(janela, text = 'Bem Vindo ao menu da Transportadora Dely')
    welcome.grid(column=0,row=0)
    option1 = Button(janela, text="Aperte para consultar trechos e modalidades",command=Consulta)
    option1.grid(column=0,row=1)
    option2 = Button(janela, text="Aperte para cadastrar um transporte" ,command=Cadastrar)
    option2.grid(column=0,row=2)
    option3 = Button(janela, text="Aperte para ver as estatísticas" ,command=Stats)
    option3.grid(column=0,row=3)
    
    quit = Button(janela, text="Aperte para sair do programa",command=janela.destroy)
    quit.grid(column=0,row=4)
    
    

########Valores iniciais
celular = ItemInfo('celular',0.5)
geladeira = ItemInfo('geladeira',60)
freezer = ItemInfo('freezer',100)
cadeira = ItemInfo('cadeira',5)
luminaria = ItemInfo('luminaria',0.8)
lavadora = ItemInfo('lavadora',120)
pequeno = Caminhao('pequeno',4.87,1000)
medio = Caminhao('medio',11.92,4000)
grande = Caminhao('grande',27.44,10000)
nViagem = [1]
custoViagem = [0]
trechos=[] 
#print(celular.qntTotal)
#print(medio.custo)



########Organização do CSV
disCity = str(Path.cwd())+ '/DNIT-Distancias.csv'
pdCity = pd.read_csv(disCity)
cidades = str(pdCity.columns.values).strip('[\'\']').split(';')#Aqui eu tento consertar a formatação do csv
#Criando uma lista com os nomes das cidades, as quais, estavam todas cadastradas como uma única coluna quando
#importadas pelo Pandas


########Interface Gráfica
janela = Tk()
janela.title('Transportadora Dely')
Welcome()
janela.mainloop()

```
